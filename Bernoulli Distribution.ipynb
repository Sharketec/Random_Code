```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import math
from matplotlib.ticker import PercentFormatter
from matplotlib import colors
```

# This is a Bernoulli Distribution

```python
#This is to fix the given data we get every time.
np.random.seed(19680801)

p = np.random.uniform(size=1)
q = 1 - p

mean = p
# Determining the Mean of the distribution:
if p < 0.5:
    meadian = 0
elif p == 1 / 2:
    meadian = np.random.uniform(size=1)
else:
    meadian = 1 / 2

# Standard deviation
sigma = np.sqrt(p*q)

x = mean+sigma*np.random.randn(300)
num_bins = 20

#Best fit line, this wasnt used in the end.
y = meadian+sigma*np.random.randn(300)


fig, axs = plt.subplots(1,2, figsize=(12,8), squeeze=True, tight_layout=True)

N, bins, patches = axs[0].hist(x, num_bins)

fracs = N/N.max()
norm = colors.Normalize(fracs.min(), fracs.max())

for thisfrac, thispatch, in zip(fracs, patches):
    color = plt.cm.viridis(norm(thisfrac))
    thispatch.set_facecolor(color)

axs[1].hist(x, num_bins, density=True, color='blue')
axs[1].yaxis.set_major_formatter(PercentFormatter(xmax=1))



plt.show()
```
